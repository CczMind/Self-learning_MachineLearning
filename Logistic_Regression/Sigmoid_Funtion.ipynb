Sigmoid 函数实现

$$g(z) = \frac{1}{1 + e^{-z}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def plot_sigmoid():
    z = np.linspace(-10, 10, 1000)
    g = sigmoid(z)

    plt.figure(figsize = (8, 6))
    plt.plot(z, g, linewidth = 2, color = "r", label = "Classic sigmoid function")
    plt.scatter([0], [0.5], color = "b", s = 50, label = "Threshold point")
    plt.title("Sigmoid Function Graph", fontsize = 15)
    plt.xlabel("z")
    plt.ylabel("g(z)")
    plt.grid(True)
    plt.legend()
    plt.axhline(y = 0.5, linewidth = 2, color = "g", linestyle = "--")
    plt.axvline(x = 0, linewidth = 2, color = "g", linestyle = "--")
    plt.text(0.1, 0.45, "(0, 0.5)", fontsize = 8)
    plt.tight_layout()
    plt.show()
```

注意当输入 z=0 时，输出为 0.5。

在逻辑回归中，这对应着分类的决策边界。

```python
plot_sigmoid()
```
